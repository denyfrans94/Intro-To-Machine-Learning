```python
#menampilkan nilai median
csv_data.median()
```

```python
#mengisi data kosong dengan median
csv_data_median = csv_data.fillna(csv_data.median())
print('Dataset yang sudah diproses Handling Missing Value dengan Mean')
print(csv_data_median.head(-1))
```

```python
#Praktek Normalisasi menggunakan Scikit Learn pada Python
import numpy as np
from sklearn import preprocessing
```

```python
array = csv_data.values

x = array[:, ([2,4,5,6,7])] #memisahkan fitur dari dataset
y = array[:, 0:2] #memisahkan class dari dataset

dataset = pd.DataFrame({
    'CustomerID':array[:,0],
    'CustomerGender':array[:,1],
    'CustomerAge':array[:,2], 
    'CustomerCity':array[:,3],
    'AnnualIncome':array[:, 4],
    'CreditScore':array[:, 5],
    'SpendingScore':array[:,6],
    'CustomerCityID':array[:,7]
})
print('Dataset sebelum di normalisasi', '\n')
print(dataset.head(10))
```

```python
#normalisasi data dengan metode MinMax

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1)) #inisialisasi normalisasi MinMax
data = min_max_scaler.fit_transform(x) #transformasi MinMax untuk fitur

dataset = pd.DataFrame({
    'CustomerAge':data[:,0],
    'AnnualIncome':data[:,1],
    'CreditScore':data[:,2],
    'SpendingScore':data[:,3],
    'CustomerCityID':data[:,4],
    'CustomerID':array[:,0],
    'CustomerGender':array[:,1],
    'CustomerCity':array[:,3]
})
print('Dataset setalah dinormalisasi')
print(dataset.head(10))
```
